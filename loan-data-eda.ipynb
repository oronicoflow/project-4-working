## Handling missing values

The features id, member_id, url and desc are completely empty - remove them. 

```python
risky_loans = risky_loans.drop(['id', 'member_id', 'url', 'desc'], axis=1, errors='ignore')
```

For categorical features (emp_title, verification_status_joint), the missing values should be filled with an empty string - put into list fill_empty.

For some of the numerical features (eg. mths_since_last_record), the missing values should be filled using the maximum value of the respective columns - put into list fill_max. 

For the rest of the numerical features (e.g. emp_length), the missing values should be filled using the minimum value of the respective columns - put into list fill_min. 

```python
fill_empty = ['emp_title', 
              'verification_status_joint']
fill_max = ['bc_open_to_buy', 
            'mo_sin_old_il_acct', 
            'mths_since_last_delinq',
            'mths_since_last_major_derog', 
            'mths_since_last_record',
            'mths_since_rcnt_il', 
            'mths_since_recent_bc', 
            'mths_since_recent_bc_dlq',
            'mths_since_recent_inq', 
            'mths_since_recent_revol_delinq',
            'pct_tl_nvr_dlq',
            'sec_app_mths_since_last_major_derog']
fill_min = np.setdiff1d(risky_loans.columns.values, np.append(fill_empty, fill_max))
```

```python
risky_loans[fill_empty] = risky_loans[fill_empty].fillna('')
risky_loans[fill_max] = risky_loans[fill_max].fillna(risky_loans[fill_max].max())
risky_loans[fill_min] = risky_loans[fill_min].fillna(risky_loans[fill_min].min())
```

## Multicollinearity

Calculate pairwise Pearson correlations for numerical features and remove features with very high correlation coefficients. 

```python
from itertools import combinations
from scipy.stats import pearsonr

num_feat = risky_loans.select_dtypes('number').columns.values
comb_num_feat = np.array(list(combinations(num_feat, 2)))
corr_num_feat = np.array([])
for comb in comb_num_feat:
    corr = pearsonr(risky_loans[comb[0]], risky_loans[comb[1]])[0]
    corr_num_feat = np.append(corr_num_feat, corr)
```

```python
high_corr_num = comb_num_feat[np.abs(corr_num_feat) >= 0.9]
high_corr_num
```

```python
# Second feature (chosen arbitrarily) removed 
risky_loans = risky_loans.drop(np.unique(high_corr_num[:, 1]), axis=1, errors='ignore')
```

```python
risky_loans.shape
```

Calculate pairwise Cramer's V correlation for categorical features and remove features with high correlation coefficients.  Note that Pearson's correlation is not applicable to categorical features and shouldn't be used.

```python
from scipy.stats import chi2_contingency

cat_feat = risky_loans.select_dtypes('object').columns.values
comb_cat_feat = np.array(list(combinations(cat_feat, 2)))
corr_cat_feat = np.array([])
for comb in comb_cat_feat:
    table = pd.pivot_table(risky_loans, values='loan_amnt', index=comb[0], columns=comb[1], aggfunc='count').fillna(0)
    corr = np.sqrt(chi2_contingency(table)[0] / (table.values.sum() * (np.min(table.shape) - 1) ) )
    corr_cat_feat = np.append(corr_cat_feat, corr)
```

```python
high_corr_cat = comb_cat_feat[corr_cat_feat >= 0.9]
high_corr_cat
```

```python
risky_loans = risky_loans.drop(np.unique(high_corr_cat[:, 1]), axis=1, errors='ignore')
```

## Target variable

The target feature for this dataset is an indicator if the loan is good (1) or bad (0). 

The description for each loan_status is provided as follows:

Current: Loan is up to date on all outstanding payments.

In Grace Period: Loan is past due but within the 15-day grace period.

Late (16-30): Loan has not been current for 16 to 30 days.

Late (31-120): Loan has not been current for 31 to 120 days.

Fully paid: Loan has been fully repaid, either at the expiration of the 3- or 5-year year term or as a result of a prepayment.

Default: Loan has not been current for an extended period of time.

Charged Off: Loan for which there is no longer a reasonable expectation of further payments.
